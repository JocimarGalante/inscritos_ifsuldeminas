#### Lista de Inscritos

```python
import pandas as pd

# Caminho do arquivo XLSX
caminho_arquivo = './Lista.xlsx'

# Lê o arquivo XLSX
tabela = pd.read_excel(caminho_arquivo)

# Exibe a tabela como HTML
from IPython.display import display, HTML
display(HTML(tabela.to_html(classes='table table-striped', index=False)))
```

```python
import matplotlib.pyplot as plt

# Filtra os dados da coluna 'Concorrência'
concorrencia_counts = tabela['Concorrência'].value_counts()

# Cria o gráfico
plt.figure(figsize=(8, 6))
concorrencia_counts.plot(kind='bar', color=['blue', 'orange'])
plt.title('Distribuição da Concorrência')
plt.xlabel('Tipo de Concorrência')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Exibe o gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Conta o total de candidatos
total_candidatos = tabela.shape[0]

# Conta os candidatos em Ampla Concorrência e nos grupos específicos
ampla_concorrencia_count = tabela['Concorrência'].value_counts().get('Ampla concorrência', 0)
grupo_especifico_count = tabela['Concorrência'].value_counts().get('Pretos/Pardos/Indígenas/Pessoa com Deficiência', 0)

# Gráfico de pizza para visualização
labels = ['Ampla Concorrência', 'Pretos/Pardos/Indígenas/Pessoa com Deficiência']
sizes = [ampla_concorrencia_count, grupo_especifico_count]
colors = ['blue', 'orange']
explode = (0.1, 0)  # "explode" o primeiro pedaço

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Exibe o gráfico
plt.title('Distribuição de Candidatos')
plt.show()
```

```python
# Número de vagas disponíveis
vagas_ampla = 28

# Calcula a probabilidade de conseguir a vaga
if ampla_concorrencia_count > 0:
    probabilidade_vaga = vagas_ampla / ampla_concorrencia_count
else:
    probabilidade_vaga = 0  # Evitar divisão por zero

# Exibe a probabilidade
print(f'Probabilidade de conseguir a vaga na Ampla concorrência: {probabilidade_vaga:.2%}')
```

```python
# Número de vagas disponíveis
vagas_ppi_pcd = 7

# Calcula a probabilidade de conseguir a vaga
if grupo_especifico_count > 0:
    probabilidade_vaga = vagas_ppi_pcd / grupo_especifico_count
else:
    probabilidade_vaga = 0  # Evitar divisão por zero

# Exibe a probabilidade
print(f'Probabilidade de conseguir a vaga na Pretos/Pardos/Indígenas/Pessoa com Deficiência: {probabilidade_vaga:.2%}')
```

```python
from IPython.display import display, HTML

# Filtra a tabela para encontrar o candidato "Jocimar Carvalho Galante"
candidato_nome = 'JOCIMAR CARVALHO GALANTE'
informacoes_candidato = tabela.loc[tabela['Nome'] == candidato_nome]

# Verifica se o candidato foi encontrado e exibe as informações
if not informacoes_candidato.empty:
    print(f'Informações de {candidato_nome}:')
    display(HTML(informacoes_candidato.to_html(index=False)))
else:
    print(f'Candidato {candidato_nome} não encontrado.')
```

```python
import matplotlib.pyplot as plt

# Conta o total de candidatos nas categorias
candidatos_ampla = tabela['Concorrência'].value_counts().get('Ampla concorrência', 0)
candidatos_grupo_especifico = tabela['Concorrência'].value_counts().get('Pretos/Pardos/Indígenas/Pessoa com Deficiência', 0)

# Dados para os gráficos
categorias = ['Ampla Concorrência', 'Pretos/Pardos/Indígenas/Pessoa com Deficiência']
quantidades = [candidatos_ampla, candidatos_grupo_especifico]

# Gráfico de Linhas
plt.figure(figsize=(10, 5))
plt.plot(categorias, quantidades, marker='o', linestyle='-', color='blue')
plt.title('Quantidade de Candidatos por Categoria')
plt.ylabel('Número de Candidatos')
plt.xticks(rotation=15)
plt.grid()
plt.show()

# Gráfico de Área
plt.figure(figsize=(10, 5))
plt.fill_between(categorias, quantidades, color='blue', alpha=0.3)
plt.plot(categorias, quantidades, marker='o', color='blue')
plt.title('Quantidade de Candidatos por Categoria')
plt.ylabel('Número de Candidatos')
plt.xticks(rotation=15)
plt.show()
```
